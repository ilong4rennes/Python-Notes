## rangeSum and exclusiveRangeSum

```python
def rangeSum(lo, hi):
    if (lo > hi):
        return 0
    else:
        return lo + rangeSum(lo+1, hi)

def testRangeSum():
    print('Testing rangeSum()...', end='')
    assert(rangeSum(3, 5) == 12) # 3+4+5
    assert(rangeSum(3, 6) == 18) # 3+4+5+6
    assert(rangeSum(6, 6) == 6)  # 6
    assert(rangeSum(7, 6) == 0)  # there are no integers in this range
    print('Passed!')
```

```python
def exclusiveRangeSum(lo, hi):
    if lo >= hi:
        return 0
    else:
        return lo + exclusiveRangeSum(lo+1, hi)

# @testFunction
def testExclusiveRangeSum():
    assert(exclusiveRangeSum(3, 5) == 7)  # 3+4 (exclude 5)
    assert(exclusiveRangeSum(3, 6) == 12) # 3+4+5 (exclude 6)
    assert(exclusiveRangeSum(5, 6) == 5)  # 5 (exclude 6)
    assert(exclusiveRangeSum(6, 6) == 0)  # there are no integers in this range
    assert(exclusiveRangeSum(7, 6) == 0)  # there are no integers in this range
```

## digitSum and maxDigit

```python
def digitSum(n):
    n = abs(n) # first make n positive if it is negative
    if (n < 10):
        return n
    else:
        onesDigit = n%10
        restOfDigits = n//10
        return onesDigit + digitSum(restOfDigits)

def testDigitSum():
    print('Testing digitSum()...', end='')
    assert(digitSum(232) == 7)  # 2+3+2
    assert(digitSum(2) == 2)    # the sum of a single digit is just itself
    assert(digitSum(-2) == 2)   # this should work for negative numbers
    assert(digitSum(-232) == 7) # 2+3+2
    print('Passed!')
```

```python
def maxDigit(n):
    n = abs(n)
    if n < 10:
        return n
    else:
        return n%10 if n%10 > maxDigit(n//10) else maxDigit(n//10)

# @testFunction
def testMaxDigit():
    assert(maxDigit(2) == 2)
    assert(maxDigit(232) == 3)
    assert(maxDigit(251340) == 5)
    assert(maxDigit(-232) == 3)
```

## fibonacci and perrin

```python
def fibonacci(n):
    if n < 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)

def testFibonacci():
    print('Testing fibonacci()...', end='')
    assert(fibonacci(0) == 1)
    assert(fibonacci(1) == 1)
    assert(fibonacci(2) == 2)
    assert(fibonacci(3) == 3)
    assert(fibonacci(4) == 5)
    assert(fibonacci(5) == 8)
    assert(fibonacci(6) == 13)
    print('Passed!')
```

```python
def perrin(n):
    if n == 0: return 3
    elif n == 1: return 0
    elif n == 2: return 2
    else:
        return perrin(n-3) + perrin(n-2)

# @testFunction
def testPerrin():
    assert(perrin(0) == 3)  # base case
    assert(perrin(1) == 0)  # base case
    assert(perrin(2) == 2)  # base case
    assert(perrin(3) == 3)  # perrin(0) + perrin(1)
    assert(perrin(4) == 2)  # perrin(1) + perrin(2)
    assert(perrin(5) == 5)  # perrin(2) + perrin(3)
    assert(perrin(6) == 5)  # perrin(3) + perrin(4)
    assert(perrin(7) == 7)  # perrin(4) + perrin(5)
    assert(perrin(8) == 10) # perrin(5) + perrin(6)
```

## reversedList and reversedEvens

```python
def reversedList(L):
    if L == []:
        return []
    else:
        first = L[0]
        rest = L[1:]
        return reversedList(rest) + [first]

def testReversedList():
    print('Testing reversedList()...', end='')
    assert(reversedList([]) == [])
    assert(reversedList([1]) == [1])
    assert(reversedList([1,2]) == [2,1])
    assert(reversedList([1,2,3]) == [3,2,1])
    assert(reversedList([1,2,3,4]) == [4,3,2,1])
    print('Passed!')
```

```python
def reversedEvens(L):
    if L == []: 
        return []
    else:
        rest = reversedEvens(L[1:])
        if L[0] % 2 == 0:
            return rest + [L[0]]
        else:
            return rest
        

# @testFunction
def testReversedEvens():
    assert(reversedEvens([]) == [])
    assert(reversedEvens([1]) == [])
    assert(reversedEvens([2]) == [2])
    assert(reversedEvens([1,2,3,4]) == [4,2])
    assert(reversedEvens([1,2,3,4,5,6,7]) == [6,4,2])
    assert(reversedEvens([1,3,5,7]) == [])
    assert(reversedEvens([2,4,6,8]) == [8,6,4,2])
```

## power and betterPower

```python
def power(base, exponent):
    if exponent == 0:
        return 1
    else:
        return base * power(base, exponent-1)

def testPower():
    print('Testing power()...', end='')
    assert(power(3, 0) == 1)
    assert(power(3, 1) == 3)
    assert(power(3, 2) == 9)
    assert(power(3, 3) == 27)
    assert(power(3, 4) == 81)
    assert(power(3, 5) == 243)
    assert(power(5, 3) == 125)
    print('Passed!')
```

```python
def betterPower(base, exponent):
    if exponent == 0:
        return 1
    else:
        if exponent > 0:
            return base * betterPower(base, exponent-1)
        else:
            exponent = abs(exponent)
            return 1 / (base * betterPower(base, exponent-1))

# @testFunction
def testBetterPower():
    assert(betterPower(2, 0) == 1)
    assert(betterPower(2, 1) == 2)
    assert(betterPower(2, 5) == 32)
    assert(betterPower(5, 2) == 25)
    assert(almostEqual(betterPower(2, -1), 1/2))
    assert(almostEqual(betterPower(2, -5), 1/32))
    assert(almostEqual(betterPower(5, -2), 1/25))
```

## Greatest Common Divisor (gcd)

```python
import time

def gcd(x, y):
    # The intuitive, though very slow, approach:
    largestDivisor = 1
    for d in range(2, min(x, y)+1):
        if ((x%d == 0) and (y%d == 0)):
            largestDivisor = d
    return largestDivisor

def testGcdTiming():
    time0 = time.time()
    x = 19**5 # 2476099
    y = 17**5 # 1419857
    result = gcd(19**5, 17**5) # this will be very slow
    time1 = time.time()
    elapsedTime = time1 - time0
    print(f'Found gcd(19**5, 17**5) == {result} in {elapsedTime} seconds')

testGcdTiming()
```

```python
import time

def gcd(x, y):
    # Euclid's (very fast) algorithm:
    if y == 0:
        return x
    else:
        return gcd(y, x%y)

def testGcdTiming():
    time0 = time.time()
    x = 19**5 # 2476099
    y = 17**5 # 1419857
    result = gcd(19**5, 17**5) # this will be very slow
    time1 = time.time()
    elapsedTime = time1 - time0
    print(f'Found gcd(19**5, 17**5) == {result} in {elapsedTime} seconds')

testGcdTiming()
```

```python
# This version uses Euclid's algorithm, and accounts
# for negative values and 0's:

def gcd(x, y):
    x, y = abs(x), abs(y)
    if (x == 0):
        return y
    elif y == 0:
        return x
    else:
        return gcd(y, x%y)

def testGcd():
    print('Testing gcd()...', end='')
    assert(gcd(50, 20) == 10)
    assert(gcd(20, 50) == 10)
    assert(gcd(500, 420) == 20)
    assert(gcd(420, 500) == 20)
    # test negatives
    assert(gcd(-420, 500) == 20)
    assert(gcd(420, -500) == 20)
    assert(gcd(-420, -500) == 20)
    # test 0's
    assert(gcd(0, -500) == 500)
    assert(gcd(-400, 0) == 400)
    print('Passed!')

testGcd()
```

## Towers of Hanoi

```python
# Towers of Hanoi with a solver!

# from cmu_graphics import *

def onAppStart(app):
    app.stepsPerSecond = 2
    resetDisks(app, 3)

def resetDisks(app, disks):
    app.disks = disks
    app.towers = [list(reversed(range(disks))), [], []]
    app.selectedTower = None
    app.moves = 0
    app.runningSolver = False

def redrawAll(app):
    drawTowers(app)
    drawLabel(f'{app.moves} Moves', 200, 350, size=16)
    drawLabel(app.message, 200, 370, size=16)

def drawTowers(app):
    baseY = 330
    baseWidth = 100
    baseHeight = 10
    postWidth = 10
    postHeight = 150
    diskHeight = 15
    for i in range(len(app.towers)):
        tower = app.towers[i]
        towerCx = 75 + 125*i
        towerColor = 'gold' if (tower == app.selectedTower) else 'red'
        # draw the tower base:
        drawRect(towerCx-baseWidth/2, baseY-baseHeight,
                 baseWidth, baseHeight, fill=towerColor)
        # draw the tower post:
        drawRect(towerCx-postWidth/2, baseY-baseHeight-postHeight,
                 postWidth, postHeight, fill=towerColor)
        # draw the disks
        diskBottomY = baseY-baseHeight
        for disk in app.towers[i]:
            diskWidth = 20 + 15*disk
            drawRect(towerCx-diskWidth/2, diskBottomY-diskHeight,
                     diskWidth, diskHeight, fill='green',
                     border='black', borderWidth=1)
            diskBottomY -= diskHeight

def onMousePress(app, mouseX, mouseY):
    if app.runningSolver:
        pass
    else:
        app.message = ''
        if mouseX < 135: tower = app.towers[0]
        elif mouseX > 265: tower = app.towers[2]
        else: tower = app.towers[1]
        if app.selectedTower == None:
            if tower == [ ]:
                app.message = 'Cannot select an empty tower'
            else:
                app.selectedTower = tower
        else:
            moveDisk(app, app.selectedTower, tower)

def moveDisk(app, fromTower, toTower):
    app.selectedTower = None
    disk = fromTower[-1]
    if (toTower == [ ]) or (disk < toTower[-1]):
        fromTower.pop()
        toTower.append(disk)
        app.moves += 1
        if (app.towers[0] == app.towers[1] == []):
            app.message = 'You did it!'
    else:
        app.message = 'Cannot move there!'

def onKeyPress(app, key):
    app.runningSolver = False
    if key.isdigit() and int(key) > 0:
        resetDisks(app, int(key))
    elif key == 'r':
        resetDisks(app, app.disks)
    elif key == 's':
        runSolver(app, app.disks)

def onStep(app):
    if app.runningSolver:
        fromIndex, toIndex = app.solutionSteps[0]
        fromTower, toTower = app.towers[fromIndex], app.towers[toIndex]
        if app.selectedTower == None:
            app.selectedTower = fromTower
        else:
            moveDisk(app, fromTower, toTower)
            app.solutionSteps.pop(0)
            if app.solutionSteps == [ ]:
                app.runningSolver = False

def runSolver(app, disks):
    resetDisks(app, disks)
    app.runningSolver = True
    app.solutionSteps = solveTowersOfHanoi(disks)

def move(n, source, target, temp):
    if n == 1:
        return [(source, target)]
    else:
        return (move(n-1, source, temp,   target) +
                move(  1, source, target, temp) +
                move(n-1, temp,   target, source))

def solveTowersOfHanoi(n):
    return move(n, 0, 2, 1)

# def main():
#     runApp()

# main()
```

Background: we just learned how to solve Towers of Hanoi using recursion. Here, we will modify our solver to return the number of long moves that the solution required, where a long move is a move between non-adjacent towers. Thus, a long move is either from the left tower to the right tower, or from the right tower to the left tower.

With this in mind, write the function hanoiLongMoveCount(n) that takes a positive integer n and returns the number of long moves in the solution found by solveTowersOfHanoi(n).

```python
def hanoiLongMoveCount(n):
    moves = moveCountHelper(n, 0, 2, 1)
    result = moves.count((0, 2)) + moves.count((2, 0))
    return result

def moveCountHelper(n, source, target, temp):
    if n == 1:
        return [(source, target)]
    else:
        return (moveCountHelper(n-1, source, temp,   target) +
                moveCountHelper(  1, source, target, temp) +
                moveCountHelper(n-1, temp,   target, source))

# @testFunction
def testHanoiLongMoveCount():
    assert(hanoiLongMoveCount(1) == 1) # [(0, 2)]
    assert(hanoiLongMoveCount(2) == 1) # [(0, 1), (0, 2), (1, 2)]
    assert(hanoiLongMoveCount(3) == 3) # [(0, 2), (0, 1), (2, 1), (0, 2),
                                       #  (1, 0), (1, 2), (0, 2)]
    assert(hanoiLongMoveCount(4) == 5) # [(0, 1), (0, 2), (1, 2), (0, 1),
                                       #  (2, 0), (2, 1), (0, 1), (0, 2),
                                       #  (1, 2), (1, 0), (2, 0), (1, 2),
                                       #  (0, 1), (0, 2), (1, 2)]
```

## Flood Fill

```python
# Flood Fill

# from cmu_graphics import *

def onAppStart(app):
    app.colors = ['red', 'orange', 'green', 'blue', 'purple', 'cyan']
    resetBoard(app)

def resetBoard(app):
    app.board = [[ 1, 1, 0, 1, 1, 2],
                 [ 0, 1, 1, 3, 1, 2],
                 [ 0, 0, 1, 3, 1, 2],
                 [ 0, 4, 4, 1, 2, 2],
                 [ 0, 4, 2, 1, 2, 5],
                 [ 0, 1, 2, 2, 2, 5]
                ]
    app.floodFillColor = 'red'

def redrawAll(app):
    drawLabel('Flood Fill', 200, 30, size=16)
    drawLabel('Press space to change the flood fill color', 200, 50, size=12)
    drawLabel('Click cell to flood fill in that color', 200, 70, size=12)
    drawLabel('Press r to reset the board', 200, 90, size=12)
    drawLabel(f'Current flood fill color: {app.floodFillColor}', 200, 110,
              size=12, fill=app.floodFillColor)
    drawBoard(app)

def drawBoard(app):
    rows,cols = len(app.board), len(app.board[0])
    for row in range(rows):
        for col in range(cols):
            drawCell(app, row, col)

def drawCell(app, row, col):
    (x0, y0, x1, y1) = getCellBounds(app, row, col)
    colorIndex = app.board[row][col]
    drawRect(x0, y0, x1-x0, y1-y0, fill=app.colors[colorIndex],
             border='black', borderWidth=1)

def getCellBounds(app, row, col):
    # Given the row,col of a cell
    # return the bounds of that cell (x0, y0, x1, y1)
    rows,cols = len(app.board), len(app.board[0])
    cellSize = 40
    gridTop = 150
    gridLeft = 200 - cellSize * cols/2
    x0 = gridLeft + col * cellSize
    x1 = gridLeft + (col+1) * cellSize
    y0 = gridTop + row * cellSize
    y1 = gridTop + (row+1) * cellSize
    return (x0, y0, x1, y1)

def getCell(app, x, y):
    # Given a mouse location (x, y),
    # return the row,col of that cell,
    # or None if it is not in a cell.
    rows,cols = len(app.board), len(app.board[0])
    cellSize = 40
    gridWidth  = cellSize * cols
    gridHeight = cellSize * rows
    gridTop = 150
    gridLeft = 200 - gridWidth/2
    gridBottom = gridTop + rows * cellSize
    gridRight = gridLeft + cols * cellSize
    if ((x < gridLeft) or (x > gridRight) or (y < gridTop) or (y > gridBottom)):
        return None
    row = int((y - gridTop) / cellSize)
    col = int((x - gridLeft) / cellSize)
    return row,col

def onKeyPress(app, key):
    if key == 'r':
        resetBoard(app)
    elif key == 'space':
        oldColorIndex = app.colors.index(app.floodFillColor)
        newColorIndex = (oldColorIndex + 1) % len(app.colors)
        app.floodFillColor = app.colors[newColorIndex]

def onMousePress(app, mouseX, mouseY):
    cell = getCell(app, mouseX, mouseY)
    if cell != None:
        row, col = cell
        floodFill(app, row, col)

def floodFill(app, row, col):
    oldValue = app.board[row][col]
    newValue = app.colors.index(app.floodFillColor)
    if oldValue != newValue:
        floodFillHelper(app.board, row, col, oldValue, newValue)

def floodFillHelper(board, row, col, oldValue, newValue):
    rows, cols = len(board), len(board[0])
    if ((row < 0) or (row >= rows) or
        (col < 0) or (col >= cols) or
        (board[row][col] != oldValue)):
        return
    else:
        board[row][col] = newValue
        floodFillHelper(board, row-1, col, oldValue, newValue) # up
        floodFillHelper(board, row+1, col, oldValue, newValue) # down
        floodFillHelper(board, row, col-1, oldValue, newValue) # left
        floodFillHelper(board, row, col+1, oldValue, newValue) # right

# def main():
#     runApp()

# main()
```

## Quick Sort

```python
def partition(L):
    # This is a non-destructive version.
    # There is a faster, though perhaps less clear, destructive version.
    pivot = L[0]
    smaller = [ ]
    larger = [ ]
    for i in range(1, len(L)):
        if L[i] <= pivot:
            smaller.append(L[i])
        else:
            larger.append(L[i])
    return smaller, pivot, larger

def testPartition():
    print('Testing partition()...', end='')
    assert(partition([3,2,5,1,4]) == ([2,1], 3, [5,4]))
    assert(partition([3,2,3,1,4]) == ([2,3,1], 3, [4]))
    print('Passed!')

testPartition()
```

```python
def quickSort(L):
    if len(L) < 2:
        return L
    else:
        smaller, pivot, larger = partition(L)
        return quickSort(smaller) + [pivot] + quickSort(larger)

def partition(L):
    pivot = L[0]
    smaller = [ ]
    larger = [ ]
    for i in range(1, len(L)):
        if L[i] <= pivot:
            smaller.append(L[i])
        else:
            larger.append(L[i])
    return smaller, pivot, larger

def testQuickSort():
    print('Testing quickSort()...', end='')
    L1 = [5,2,0,7,6,3,1,4]
    assert(quickSort(L1) == sorted(L1))
    L2 = [8,0,4,6,7,5,9,3,2,1]
    assert(quickSort(L2) == sorted(L2))
    print('Passed!')

testQuickSort()
```

## Merge Sort

```python
def merge(L, M):
    # This is a non-destructive version.
    # There is a faster, though perhaps less clear, destructive version.
    N = [ ]
    i = j = 0
    while (i < len(L)) or (j < len(M)):
        if j == len(M):
            # We used all of M, so we must use L
            useL = True
        elif i == len(L):
            # We used all of L, so we must use M
            useL = False
        else:
            # Use the smaller value from L or M
            useL = (L[i] < M[j])
        if useL:
            N.append(L[i])
            i += 1
        else:
            N.append(M[j])
            j += 1
    return N

def testMerge():
    print('Testing merge()...', end='')
    assert(merge([2,5], [1,6]) == [1,2,5,6])
    assert(merge([1,2], [5,6]) == [1,2,5,6])
    assert(merge([1,2,6], [5]) == [1,2,5,6])
    assert(merge([1,2,5,6], []) == [1,2,5,6])
    print('Passed!')

testMerge()
```
